```python
# Import libraries
import pandas as pd
```

```python
# Load CSV
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')
```

```python
# 1. Handling Missing Values

# Fill specific column
df['ADDRESSLINE2'] = df['ADDRESSLINE2'].fillna('')

# Drop remaining rows with missing values (optional â€“ adjust if needed)
df = df.dropna()
```

```python
# 2. Remove Duplicates
df = df.drop_duplicates()
```

```python
# 3. Standardize Column Names
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# 4. Standardize Text Columns
if 'country' in df.columns:
    df['country'] = df['country'].str.strip().str.title()
```

```python
# 5. Fix Date Format
if 'orderdate' in df.columns:
    df['orderdate'] = pd.to_datetime(df['orderdate'], errors='coerce')
    df['formatted_date'] = df['orderdate'].dt.strftime('%d-%m-%Y')
```

```python
# 6. Fix Data Types
if 'quantityordered' in df.columns:
    df['quantityordered'] = pd.to_numeric(df['quantityordered'], errors='coerce')
```

```python
# 7. Final Check
print("Final Data Types:\n", df.dtypes)
print("\nSummary Statistics:\n", df.describe())
```

```python
# 8. Save Cleaned File
output_file = "cleaned_sales_data.csv"
df.to_csv(output_file, index=False)

print(f"\nCleaned file saved as: {output_file}")
```
